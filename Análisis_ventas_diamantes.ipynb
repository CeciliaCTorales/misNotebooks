```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


diamantes = sns.load_dataset("diamonds")
#Visualizar primeras filas
diamantes.columns
```

```python
# Resumen estadístico de los datos
diamantes["price"].describe()
```

```python
# La mediana es igual al 2do. Quartil
np.median(diamantes["price"])
```

Resúmen estadístico de los datos
- Los pesos de los diamantes(carat) oscilan entre 0.20 y 5.01 con un media de 0.79
- Los precios de los diamantes(price) oscilan entre 326 y 18823 con un promedio de 3932



```python
#Crear un boxplot que muestre  cómo se distribuyen los valores de la variable numérica 'price' sobre la variable categórica 'color'
sns.boxplot(y='price', x='color', data= diamantes, palette = 'pastel')
plt.show()
```

```python

# Crear un gráfico de distribución de los precios con Seaborn
# Ver cómo agregar líneas verticales de referencia
plt.figure(figsize=(15,6))
sns.histplot(data=diamantes, x="price", kde=True)
#calcular la media y el desvío
media = diamantes["price"].mean()
desvio = diamantes["price"].std()
mediana = diamantes["price"].median()

# Agregar líneas verticales
plt.axvline(media, color='magenta', linestyle='--',  label = f'Media: {media:.2f}')
plt.axvline(mediana, color='deeppink', linestyle='--',  label = f'Mediana: {mediana:.2f}')
plt.axvline(desvio, color='cyan', linestyle='--',  label = f'Desvío STD: {desvio:.2f}')

# Agregar líneas verticales de mean +- 1 desvío 68%
plt.axvline(media + desvio, color= "blue", linestyle='--', label= f'Media + 1 desvio(68%): {media + desvio:.2f}')
plt.axvline(media - desvio, color= "blue", linestyle='--', label= f'Media - 1 desvio(68%): {media - desvio:.2f}')

# Agregar líneas verticales de mean +- 2 desvíos 95%
plt.axvline(media + (desvio*2), color= "orange", linestyle='--', label= f'Media + 2 desvio(95%): {media + (desvio*2):.2f}')
plt.axvline(media - (desvio*2), color= "orange", linestyle='--', label= f'Media - 2 desvio(95%): {media - (desvio*2):.2f}')

# Agregar líneas verticales de mean +- 3 desvíos 99,7%
plt.axvline(media + (desvio*3), color= "pink", linestyle='--', label= f'Media + 3 desvio(99,7%): {media + (desvio*3):.2f}')
plt.axvline(media - (desvio*3), color= "pink", linestyle='--', label= f'Media - 3 desvio(99,7%): {media - (desvio*3):.2f}')

plt.legend()
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.title("Distribución de Precios con Seaborn")
plt.show()
```
