```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

diamantes = sns.load_dataset("diamonds")
#Visualizar primeras filas
diamantes.columns
```

```python
# Resumen estadístico de los datos
diamantes.describe()
```

Resúmen estadístico de los datos
- Los pesos de los diamantes(carat) oscilan entre 0.20 y 5.01 con un media de 0.79
- Los precios de los diamantes(price) oscilan entre 326 y 18823 con un promedio de 3932



```python
#Gráficos de distribución
plt.figure(figsize=(12,5))



# Visualizar la distribución de los precios de los diamantes
plt.subplot(1,2,1)
sns.histplot(diamantes["price"], bins=30, kde=True)
plt.title("Distribución de los precios")
plt.xlabel("Precios")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

```python
# Visualizar distribución de los precios y pesos con Boxplot

plt.figure(figsize=(12,5))

# Distribución de pesos
plt.subplot(1,2,1)
sns.boxenplot(diamantes["carat"])
plt.title("Distribución de Pesos")

# Distribución de precios
plt.subplot(1,2,2)
sns.boxplot(diamantes['price'])
plt.title("Distribución de precios")

plt.show()

```
